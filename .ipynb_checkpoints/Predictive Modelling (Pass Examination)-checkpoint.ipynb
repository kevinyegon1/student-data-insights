```python
import pandas as pd
df = pd.read_csv(r"C:\Users\hi\Desktop\Data Analytics\Data Sets\new\python_learning_exam_performance.csv")
df = df.dropna().reset_index()
df = df.drop(columns=['index'])
df.info()
```

```python
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier, VotingClassifier
from xgboost import XGBClassifier
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
y = df['passed_exam']
X = df.drop(columns=['student_id','final_exam_score','passed_exam'])
```

```python
#identifying column types
num_col = X.select_dtypes('int64','float64').columns
cat_col = X.select_dtypes('object').columns
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.25)
```

```python
# preprocessing
preprocessing = ColumnTransformer(
    transformers=[
        ('num',StandardScaler(), num_col),
        ('cat',OneHotEncoder(handle_unknown='ignore'), cat_col)
    ]
)
```

```python
# defining base models
rfc = RandomForestClassifier(random_state=42, n_estimators=100,class_weight='balanced') 

xgb = XGBClassifier(random_state=42)
```

```python
# ensembling using votting classifier
ensemble = VotingClassifier(
    estimators=[('rf',rfc),('xgb',xgb)],
    voting='soft',
)
```

```python
# imblearn pipeline
model = Pipeline(
    steps=[
        ('preprocessing',preprocessing),
        ('smote',SMOTE(random_state=42)),
        ('classifier', ensemble)
    ]
)
```

```python
#model training and prediction
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
print(classification_report(y_test, y_pred))
```

```python
from sklearn.model_selection import GridSearchCV, StratifiedKFold

# 1. Define the Parameter Grid with correct 'double underscore' paths
# Syntax: [step_name]__[parameter] OR [step_name]__[model_name]__[parameter]
param_grid = {
    # Adjust the SMOTE ratio
    'smote__sampling_strategy': [0.7, 1.0],
    
    # Adjust the Ensemble voting type
    'classifier__voting': ['soft'], # Soft is usually better for probabilities
    
    # Tune the Random Forest inside the ensemble
    'classifier__rf__n_estimators': [100, 200],
    'classifier__rf__max_depth': [None, 10],
    
    # Tune the XGBoost inside the ensemble
    'classifier__xgb__learning_rate': [0.01, 0.1],
    'classifier__xgb__n_estimators': [100],
    
    # Balance the influence of RF vs XGB
    'classifier__weights': [[1, 1], [2, 1], [1, 2]]
}

# 2. Setup Cross-Validation
cv_strategy = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# 3. Initialize GridSearch
# We use 'f1' to balance Precision and Recall for the failing students
grid_search = GridSearchCV(
    estimator=model, 
    param_grid=param_grid, 
    cv=cv_strategy, 
    scoring='f1', 
    n_jobs=-1,
    verbose=1
)

# 4. Fit the model
grid_search.fit(X_train, y_train)

# 5. Output Results
print(f"Best Score (F1): {grid_search.best_score_:.4f}")
print("Best Parameters found:")
for param, value in grid_search.best_params_.items():
    print(f"{param}: {value}")
```
